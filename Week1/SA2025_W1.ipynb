### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
df = pd.read_csv('Cars.csv')

median_mpg = df['mpg'].median()
ab_median = df[df['mpg'] > median_mpg].copy()

ab_median['hp_to_wt'] = ab_median['horsepower'] / ab_median['weight']
best_ratio = ab_median[ab_median['hp_to_wt'] == ab_median['hp_to_wt'].max()][['name', 'hp_to_wt']]

print(best_ratio)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

sns.set(style = "whitegrid")

plt.figure(figsize = (12,6))
sns.lineplot(data = mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title = 'Origin')
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
df = df.dropna(subset=['horsepower'])
hue_order = ['japan', 'europe', 'usa']

sns.set(style = "whitegrid")

plt.figure(figsize = (12,6))
scatter = sns.scatterplot(
    data = df,
    x = 'horsepower',
    y = 'weight',
    hue = 'origin',
    size = 'mpg',
    palette = 'deep',
    hue_order = hue_order,
    sizes = (20,200),
    alpha = 0.7
)

plt.title('Car Horsepower vs Weight by Origin and MPG', fontsize = 16)
plt.xlabel('Horsepower', fontsize = 12)
plt.ylabel('Weight', fontsize = 12)
plt.legend(title = 'Origin', bbox_to_anchor = (1,1))
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
name_years = df.groupby('name')['model_year'].nunique()
multi_year_names = name_years[name_years > 1].index

multi_year_df = df[df['name'].isin(multi_year_names)]

mpg_variation = multi_year_df.groupby('name').agg(
    appearances=('mpg', 'count'),
    std_mpg=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()

consistent_models = mpg_variation[mpg_variation['std_mpg'] < 1.0]
consistent_models_sorted = consistent_models.sort_values(
    by=['appearances', 'avg_mpg'],
    ascending=[False, False]
)
consistent_models_sorted
```
